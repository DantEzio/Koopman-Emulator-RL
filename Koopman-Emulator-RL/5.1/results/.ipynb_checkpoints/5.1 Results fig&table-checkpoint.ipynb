```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#MSE and NSE
def MSE(x,y):
    return np.sqrt(np.mean(np.square(x-y)))

def NSE(pre,tar):
    a=np.sum(np.square(pre-tar))/np.sum(np.square(tar-np.mean(tar)))
    return 1-a
```

# Results of Emulation

## Get results data

```python
modelist=['DLEDMD','Linear','MLP','SWMM']
num=[0,1,2,3]
dic=[]
for m in modelist:
    for n in num:
        dic.append(m+str(n))
```

```python
pred,reg={},{}
#pred
for it in dic:
    pred[it]=pd.read_excel('./Prediction/'+it+'.xlsx').values[:-1]
    reg[it]=pd.read_excel('./Regression/'+it+'.xlsx').values[:-1]
```

## Prediction

```python
#Prediction 4 rainfall for testing
#Prediction
results=[]
for i in range(4):
    print('Rain number:',i)
    Y_test=pred['SWMM'+str(i)]
    
    print('DLEDMD:')
    y_D=pred['DLEDMD'+str(i)]
    print(y_D.shape,Y_test.shape)
    print('MSE:',MSE(y_D,Y_test[:-1]),'  NSE:',NSE(y_D,Y_test[:-1]))
    results.append([MSE(y_D,Y_test[:-1]),NSE(y_D,Y_test[:-1])])
    
    print('Linear:')
    y_L=pred['Linear'+str(i)]
    print('MSE:',MSE(y_L,Y_test),'  NSE:',NSE(y_L,Y_test))
    results.append([MSE(y_L,Y_test),NSE(y_L,Y_test)])
    
    print('MLP:')
    y_M=pred['MLP'+str(i)]
    print('MSE:',MSE(y_M,Y_test[:-1]),'  NSE:',NSE(y_M,Y_test[:-1]))
    results.append([MSE(y_M,Y_test[:-1]),NSE(y_M,Y_test[:-1])])
    
pd.DataFrame(results).to_excel('pre_result.xlsx')
```

## Regression

```python
#Prediction 4 rainfall for testing
#Prediction
results=[]
for i in range(4):
    print('Rain number:',i)
    
    Y_test=reg['SWMM'+str(i)]
    
    print('DLEDMD:')
    y_D=reg['DLEDMD'+str(i)]
    print('MSE:',MSE(y_D,Y_test[:-1]),'  NSE:',NSE(y_D,Y_test[:-1]))
    results.append([MSE(y_D,Y_test[:-1]),NSE(y_D,Y_test[:-1])])
    
    print('Linear:')
    y_L=reg['Linear'+str(i)]
    print('MSE:',MSE(y_L,Y_test),'  NSE:',NSE(y_L,Y_test))
    results.append([MSE(y_L,Y_test),NSE(y_L,Y_test)])
    
    print('MLP:')
    y_M=reg['MLP'+str(i)]
    print('MSE:',MSE(y_M,Y_test[:-1]),'  NSE:',NSE(y_M,Y_test[:-1]))
    results.append([MSE(y_M,Y_test[:-1]),NSE(y_M,Y_test[:-1])])
    
    
pd.DataFrame(results).to_excel('reg_result.xlsx')
```
