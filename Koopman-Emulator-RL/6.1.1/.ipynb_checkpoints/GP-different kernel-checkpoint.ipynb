```python
#MSE and NSE
def MSE(x,y):
    return np.sqrt(np.mean(np.square(x-y)))

def NSE(pre,tar):
    a=np.sum(np.square(pre-tar))/np.sum(np.square(tar-np.mean(tar)))
    return 1-a
```

```python
def Pre_data():
    #Prepare training data (set0)
    df_s0 = pd.read_excel(r'./save_data/excelfile/rain0_set0_state_data.xlsx')
    df_a0= pd.read_excel(r'./save_data/excelfile/rain0_set0_action_data.xlsx')
    df_r0= pd.read_excel(r'./save_data/excelfile/rain0_set0_rain_data.xlsx')
    dt_s0=df_s0.values[:1,1:]
    dt_a0=df_a0.values[:1,1:]
    dt_r0=df_r0.values[:1,1:]
    
    data=np.concatenate((dt_s0,dt_a0,dt_r0),axis=1)
    for it in [0,2,4,6,8,9]:
        for jt in [0,1,2]:   
            df_s0 = pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_state_data.xlsx')
            df_a0= pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_action_data.xlsx')
            df_r0= pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_rain_data.xlsx')
            dt_s0=df_s0.values[:,1:]
            dt_a0=df_a0.values[:,1:]
            dt_r0=df_r0.values[:,1:]
            tem=np.concatenate((dt_s0,dt_a0,dt_r0),axis=1)
            data=np.concatenate((data,tem),axis=0)
    
    data=data[1:,:]
            
    #Prepare test1 data
    df_s0 = pd.read_excel(r'./save_data/excelfile/rain0_set1_state_data.xlsx')
    df_a0= pd.read_excel(r'./save_data/excelfile/rain0_set1_action_data.xlsx')
    df_r0= pd.read_excel(r'./save_data/excelfile/rain0_set1_rain_data.xlsx')
    dt_s0=df_s0.values[:,1:]
    dt_a0=df_a0.values[:,1:]
    dt_r0=df_r0.values[:,1:]
    data_test1=[]
    data_test1.append(np.concatenate((dt_s0,dt_a0,dt_r0),axis=1))
    for it in [3,5,7]:
        for jt in [1]:   
            df_s0 = pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_state_data.xlsx')
            df_a0= pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_action_data.xlsx')
            df_r0= pd.read_excel(r'./save_data/excelfile/rain'+str(it)+'_set'+str(jt)+'_rain_data.xlsx')
            dt_s0=df_s0.values[:,1:]
            dt_a0=df_a0.values[:,1:]
            dt_r0=df_r0.values[:,1:]
            tem=np.concatenate((dt_s0,dt_a0,dt_r0),axis=1)
            data_test1.append(tem)
    
    #Prepare training data2 (set0)
    df_s0 = pd.read_excel(r'./save_data/excelfile/rain0_set0_state_data.xlsx')
    df_a0= pd.read_excel(r'./save_data/excelfile/rain0_set0_action_data.xlsx')
    df_r0= pd.read_excel(r'./save_data/excelfile/rain0_set0_rain_data.xlsx')
    dt_s0=df_s0.values[:,1:]
    dt_a0=df_a0.values[:,1:]
    dt_r0=df_r0.values[:,1:]
    
    return data,data_test1

data,data_test1=Pre_data()
print(data.shape)
```

# GPR

## Dotkernel

```python
X_train=data[:data.shape[0]-2,:]
Y_train=data[1:data.shape[0]-1,:4]

kernel=DotProduct()

#GPR
gpr=GaussianProcessRegressor(kernel=kernel,random_state=0)
gpr.fit(X_train,Y_train)
```

```python
X_train=data[:data.shape[0]-2,:]
Y_train=data[1:data.shape[0]-1,:4]

kernel=RBF()

#GPR
gprw=GaussianProcessRegressor(kernel=kernel,random_state=0)
gprw.fit(X_train,Y_train)
```

```python
X_train=data[:data.shape[0]-2,:]
Y_train=data[1:data.shape[0]-1,:4]

kernel=WhiteKernel()+DotProduct()

#GPR
gprC=GaussianProcessRegressor(kernel=kernel,random_state=0)
gprC.fit(X_train,Y_train)
```

# Results of Emulation

## Prediction

```python
#Prediction 4 rainfall for testing
#Prediction
GPR,GPRW,GPRC,SWMM=[],[],[],[]
for i in range(4):
    #X_test, Y_test=data_test1[rainlog][:46,:],data_test1[0][1:47,:]
    X_test=data_test1[i][:47,:]
    Y_test=data_test1[i][1:48,:4]
    
    print('Rain number:',i)
    
    print('GPD:')
    y_L=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gpr.predict(X[t].reshape((1,-1)))  
        y_L[t,:]=ytem
        if t+1<Y_test.shape[0]:
            X[t+1,:4]=ytem
    print('MSE:',MSE(y_L,Y_test),'  NSE:',NSE(y_L,Y_test))
    GPR.append(y_L)
    
    print('GPW:')
    y_G=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gprw.predict(X[t].reshape((1,-1)))  
        y_G[t,:]=ytem
        if t+1<Y_test.shape[0]:
            X[t+1,:4]=ytem
    print('MSE:',MSE(y_G,Y_test),'  NSE:',NSE(y_G,Y_test))
    GPRW.append(y_G)
    
    print('GPW+D:')
    y_M=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gprC.predict(X[t].reshape((1,-1)))  
        y_M[t,:]=ytem
        if t+1<Y_test.shape[0]:
            X[t+1,:4]=ytem
    print('MSE:',MSE(y_M,Y_test),'  NSE:',NSE(y_M,Y_test))
    GPRC.append(y_M)
    
    SWMM.append(Y_test)
    
```

## Regression

```python
#Prediction 4 rainfall for testing
#Regression
GPRR,GPRWR,GPRCR,SWMMR=[],[],[],[]
for i in range(4):
    #X_test, Y_test=data_test1[rainlog][:46,:],data_test1[0][1:47,:]
    X_test=data_test1[i][:47,:]
    Y_test=data_test1[i][1:48,:4]
    
    print('Rain number:',i)
    
    print('GPr:')
    y_L=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gpr.predict(X[t].reshape((1,-1)))  
        y_L[t,:]=ytem
        
    print('MSE:',MSE(y_L,Y_test),'  NSE:',NSE(y_L,Y_test))
    GPRR.append(y_L)
    
    print('GPw:')
    y_G=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gprw.predict(X[t].reshape((1,-1)))  
        y_G[t,:]=ytem
        
    print('MSE:',MSE(y_G,Y_test),'  NSE:',NSE(y_G,Y_test))
    GPRWR.append(y_G)
    
    print('GPRC:')
    y_M=np.zeros(Y_test.shape)
    X=X_test.copy()
    for t in np.arange(Y_test.shape[0]):
        ytem=gprC.predict(X[t].reshape((1,-1)))  
        y_M[t,:]=ytem
        
    print('MSE:',MSE(y_M,Y_test),'  NSE:',NSE(y_M,Y_test))
    GPRCR.append(y_M)
    
    SWMMR.append(Y_test)
```
